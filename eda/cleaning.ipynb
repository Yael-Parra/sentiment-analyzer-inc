```python
def read_csv_from_drive(drive_url):
    """
    Download and load a CSV file from Google Drive without saving it locally.
    """
    # Extraer file ID de la URL
    file_id = drive_url.split("/d/")[1].split("/")[0]
    download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

    response = requests.get(download_url)
    response.raise_for_status()

    # Leer CSV directamente desde memoria
    df = pd.read_csv(BytesIO(response.content))
    return df

# Uso
drive_url = "https://drive.google.com/file/d/1bG7fA273jIBgJfc6YS1vsKfr1qRiNUTU/view"
df_raw_1 = read_csv_from_drive(drive_url)
```

#### <span style="color:#F3A006"> 1️⃣ Columns </span>

- Column's names format: lowercase with underscore as spaces

```python
# Changin the names to lowercase and with a underscore to as spaces to make it easier to work with.
df_raw_1.columns = (
    df_raw_1.columns
    .str.replace(r'(?<!^)(?=[A-Z])', '_', regex=True)  # Insert underscore before each uppercase (except start)
    .str.lower()                                       # Convert all to lowercase
    .str.strip()                                       # Remove leading and trailing spaces in case there were any
)
```

#### <span style="color:#F3A006"> 2️⃣ Duplicates </span>

- There are no duplicate columns

```python
# Checking if there are duplicate columns
df_raw_1.columns.duplicated().any()        # np.False_
df_raw_1["comment_id"].duplicated().any()  # np.False_
df_raw_1["comment_id"].nunique()           # 1,000
df_raw_1["video_id"].duplicated().any()    # np.True_
df_raw_1["video_id"].nunique()             # 13
```

#### <span style="color:#F3A006"> 3️⃣ Nulls </span>

- There are no nulls

```python
# Checking if there are nulls
df_raw_1.isnull().sum()
```

## <span style="color:#F3A006"> 3️⃣ Eliminating invisible row breakes </span>

```python
# Eliminating row breaks and extra spaces in the comments
df_raw_1["text"] = df_raw_1["text"].str.replace(r'\s+', ' ', regex=True).str.strip()
```
